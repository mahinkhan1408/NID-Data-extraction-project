```python
import pandas as pd
import matplotlib.pyplot as plt
class RetailSalesAnalyzer:
  def __init__(self):
    self.data = pd.read_csv("/content/retail_sales.csv")
    self.data['Date'] =pd.to_datetime(self.data['Date'])

  def data_clean(self):
    self.data.dropna(inplace = True)
  def totalsales(self):
    return self.data.groupby('Product')['Sales'].sum()
  def bestselling(self):
    return self.data.groupby('Product')['Sales'].sum().sort_values(ascending = False).index[0]
  def avrgsell(self):
    self.data.groupby('Date')['Sales'].mean()
  def plottrend(self):
    self.data.groupby('Date')['Sales'].sum().plot(kind='line')
    plt.title('sales trend over time')
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.show()

  def salesperprod(self):
    self.totalsales().plot(kind='bar')

    plt.xlabel("Product")
    plt.title("Sales per prod")
    plt.ylabel("Total Sales")
    plt.show()


```

```python
analyzer = RetailSalesAnalyzer()
print('total sales per prod \n',analyzer.totalsales() )
print("Best selling \n", analyzer.bestselling() )
print("AVg sales \n", analyzer.avrgsell())

analyzer.salesperprod()
analyzer.plottrend()
```
