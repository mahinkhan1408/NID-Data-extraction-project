```python
import pandas as pd

import matplotlib.pyplot as plt

data = {'Date':['2024-08-01','2024-08-02','2024-08-03'], 'Price':[100,50,400]}

df= pd.DataFrame(data)

df['Date']= pd.to_datetime(df['Date'])
df.plot(x='Date', y='Price',kind='line')
plt.xlabel('Data')
plt.ylabel('Price')

plt.show()
```

```python
data = {'prod':['a','b','c'],'Sales':[100,350,180]}

df= pd.DataFrame(data)

df.plot(x= 'prod', y='Sales', kind='bar')
plt.show()
```

```python
data = {'prod':['a','b','c'],'Sales':[100,350,180]}

df= pd.DataFrame(data)

df.plot(x='prod', y='Sales', kind = 'pie')

plt.show()
```
